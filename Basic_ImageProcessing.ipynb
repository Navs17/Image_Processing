```python
from skimage import io
img = io.imread("puppy.jpg")
io.imshow(img)
```

```python
img.shape
```

```python
import numpy as np
np.save("pixel_values.npy",img)
img_load = np.load("pixel_values.npy")
img_load
```

```python
img_load[3599,2399]
```

```python
from skimage import color, data
from pylab import *

#Convert to HSV
img_hsv = color.rgb2hsv(img)

#Convert back to RGB
img_rgb = color.hsv2rgb(img_hsv)

io.imshow(img_hsv)
```

```python
io.imshow(img_rgb)
```

```python
img_lab = color.rgb2lab(img)
io.imshow(img_lab)
```

```python
img_yuv = color.rgb2yuv(img)
io.imshow(img_yuv)
```

```python
img_ypbpr = color.rgb2ypbpr(img)
io.imshow(img_ypbpr)
```

```python
#Creating Basic Drawing
from skimage import draw

img_gray = color.rgb2gray(img)
x,y = draw.line(0,0,1500,1000)
img_gray[x,y] = 1
io.imshow(img_gray)
```

```python
x,y = draw.disk((500,500),100)
io.imshow(img_gray)
```

```python
from skimage.transform import rotate, resize
img_rot = rotate(img, 20)
io.imshow(img_rot)
```

```python
img_res = resize(img, (100,100))
io.imshow(img_res)
```
